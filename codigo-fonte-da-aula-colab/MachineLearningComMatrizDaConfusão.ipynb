Precision da matriz da confusão

```python
from sklearn.metrics import precision_score
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
```

```python
y_true = [0, 1, 1, 0, 1, 0, 1, 0, 0, 1]
y_pred = [0, 0, 1, 0, 1, 0, 1, 1, 0, 1]
```

```python
precisão = precision_score(y_true, y_pred)
```

```python
print(f"O numero de precisão é: {precisão}")
```

```python
#matriz da confusão
matriz = confusion_matrix(y_true, y_pred)
display = ConfusionMatrixDisplay(matriz, display_labels=[0,1])
display.plot(cmap=plt.cm.Blues, values_format='.3g')
plt.show()
```

Precision, Reacll, F1 score

```python
from sklearn.metrics import precision_score, recall_score, f1_score
```

```python
y_true = [0, 1, 1, 0, 1, 0, 1, 0, 0, 1]
y_pred = [0, 0, 1, 0, 1, 0, 1, 1, 0, 1]
```

```python
precisão = precision_score(y_true,y_pred)

lembrar = recall_score(y_true,y_pred)
pontuação = f1_score(y_true,y_pred)
```

```python
print(f"O numero de precisão é: {precisão}")
print(f"O numero de lembrar é: {lembrar}")
print(f"O numero de pontuação é: {pontuação}")
```
