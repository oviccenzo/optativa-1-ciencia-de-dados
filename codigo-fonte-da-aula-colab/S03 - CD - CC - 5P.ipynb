```python
import pandas as pd
```

```python
df = pd.read_csv('https://media.geeksforgeeks.org/wp-content/uploads/nba.csv')
df
```

```python
df.dropna(subset=['Height', 'Weight'], inplace=True)
```

```python
def convertWeight(x):
  return x * 0.453592
```

```python
def convertHeight(x):
  foot = int(x.split('-')[0])
  inch = int(x.split('-')[1])
  return ((foot*12) + inch) * 0.0254 # convertendo tudo para metros
```

```python
# Compreensão de listas
df['Weight (kg)'] = [convertWeight(x) for x in df['Weight']]
df['Height (m)']  = [convertHeight(x) for x in df['Height']]
```

```python
colunas = ['Name', 'Team', 'Position', 'Age', 'Height', 'Height (m)', 'Weight', 'Weight (kg)', 'College', 'Salary']
df = df[colunas]
```

```python
df
```

```python
# Groupby
grupos = df.groupby(by='Position')
```

```python
# Grupos
grupos.get_group('Kansas')['Salary'].mean()
```
