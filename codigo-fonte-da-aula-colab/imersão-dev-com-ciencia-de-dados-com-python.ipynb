```python
traducao_senioridade = {
    'SE':  'Senior',
    'MI':  'Pleno',
    'EN':  'Junior',
    'EX':  'Executivo'
}
df['senioridade'] = df['senioridade'].replace(traducao_senioridade)
display(df['senioridade'].value_counts())
```

```python
traducao_contrato = {
    'FT': 'Tempo Integral',
    'CT': 'Contrato',
    'PT': 'Tempo Parcial',
    'FL': 'Freelancer'
}
df['contrato'] = df['contrato'].replace(traducao_contrato)
display(df['contrato'].value_counts())
```

```python
traducao_tamanho_empresa = {
    'M': 'Médio',
    'L': 'Grande',
    'S': 'Pequeno'
}
df['tamanho_da_empresa'] = df['tamanho_da_empresa'].replace(traducao_tamanho_empresa)
display(df['tamanho_da_empresa'].value_counts())
```

```python
traducao_remoto = {
    0: 'Presencial',
    50: 'Híbrido',
    100: 'Remoto'
}
df['remoto'] = df['remoto'].replace(traducao_remoto)
display(df['remoto'].value_counts())
```

```python
import numpy as np
import pandas as pd

#criação de um dataframe de teste para usar de exemplo
df_salarios = pd.DataFrame({
    'nome' : ["Ana","Bruno","Carlos","Daniele","Val"],
    'moeada' : [4000,np.nan, 5000,np.nan,100000]
})
#Calcular a media salarial e substitui os nulos pelo media e arrendonda os valores
df_salarios['salario_media'] = df_salarios['moeada'].fillna(df_salarios['moeada'].mean().round(2))

'''Calcula mediana e substitui os nulos pela mediana'''

df_salarios['salario_mediana'] = df_salarios['moeada'].fillna(df_salarios['moeada'].median())

display(df_salarios)
```

```python
df_temperatura = pd.DataFrame({
    "Dia":["segunda" , "terca", "quarta","quinta","sexta"],
    "temperatura":[30,np.nan, np.nan,28,27]
})
df_temperatura["Preenchido_ffill"] = df_temperatura["temperatura"].ffill()
```

```python
import pycountry as pc
```

```python
def iso2_to_iso3(code):
  try:
    return pc.countries.get(alpha_2=code).alpha_3
  except:
    return None
```

```python
df_limpo['residencia_iso3'] = df_limpo['residencia'].apply(iso2_to_iso3)
```

```python
df_ds = df_limpo[df_limpo['cargo'] == 'Data Scientist']
media_ds_pais = df_ds.groupby('residencia_iso3')['usd'].mean().reset_index()
```

```python
fig = px.choropleth(media_ds_pais,
                    locations='residencia_iso3',
                    color='usd',
                    color_continuous_scale='rdylgn',
                    title='Salário médio de Cientista de Dados por país',
                    labels={'usd': 'Salário médio (USD)', 'residencia_iso3': 'País'})

fig.show()
```
